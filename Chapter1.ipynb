#### 클래스

- 지금까지 int와 str 등의 자료형을 알아봄
- 그리고 type() 함수로 원하는 데이터의 자료형을 알아낼 수 있음

- 이들은 내장된 자료형 > 파이썬이 기본으로 제공하는 자료형임

- 새로운 클래스를 정의

- 개발자가 직접 클래스를 정의하면 독자적인 자료형을 만들 수 있음
- 클래스에는 그 클래스만의 전용 함수(메서드)와 속성을 정의 가능

- 파이썬에서는 class라는 키워드를 사용하여 클래스를 정의

```python
# 클래스 구조
class 클래스 이름:
    def __init__(self, 인수, ....): # 생성자
        ...
    def 메서드 이름 1(self, 인수, ....) # 메서드 1
        ...
    def 메서드 이름 2(self, 인수, ...) # 메서드 2
        ...
```

- 클래스 정의에는 __init__라는 특별한 메서드가 있음

- 클래스를 초기화하는 방법을 정의
- 이 초기화용 메서드를 생성자(constructor)라고도 함
- 클래스의 인스턴스가 만들어질 때 한 번만 불림
- 파이썬에서는 메서드의 첫 번째 인수로 자신(자신의 인스턴스)을 나타내는 self를 명시적으로 쓰는 것이 특징

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello" + self.name + "!")

    def goodbye(self):
        print("Good-bye" + self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

- Man이라는 새로운 클래스를 정의

- Man 클래스에서 m이라는 인스턴스(객체)를 생성
- Man의 생성자(초기화 메서드)는 name이라는 인수를 받음
- 그 인수로 인스턴스 변수인 self.name을 초기화

- 인스턴스 번수 : 인스턴스별로 저장되는 변수

- 파이썬에서는 self.name처럼 self 다음에 속성 이름을 써서 인스턴스 변수를 작성하거나 접근할 수 있음

### 넘파이

- 딥러닝을 구현하다 보면 배열이나 행렬 계산이 많이 등장

- 넘파이의 배열 클래스인 numpy.array애는 편리한 메서드가 많이 준비됨
- 딥러닝을 구현할 때 이 메서드들을 이용

#### 넘파이 가져오기

- 넘파이는 외부 라이브러리
- 여기서 말하는 '외부'는 표준 파이썬에는 포함되지 않는다는 것

- 넘파이 라이브러리를 쓸 수 있도록 가져와야(import) 함

```python
import numpy as np
```

#### 넘파이 배열 생성하기

- 넘파이 배열을 만들 때는 np.array() 메서드를 이용

- np.array()는 파이썬의 리스트를 인수로 받아 넘파이 라이브러리가 제공하는 특수한 형태의 배열(numpy.ndarray)을 반환

```python
x = np.array([1.0, 2.0, 3.0])
print(x)
```

```python
type(x)
```

#### 넘파이의 산술 연선

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
x + y
```

```python
x - y
```

```python
x * y
```

```python
x / y
```

- x와 y의 원소 수가 같다면 산술 연산은 각 원소에 대해서 행해짐

- 원소 수가 다르면 오류 발생
- 원소별 > element-wise
- 원소별 곱셈 > element-wise product

- 넘파이 배열은 원소별 계산뿐 아니라 넘파이 배열과 수치 하나(스칼라값)의 조합으로 된 산술 연산도 수행 가능

- 이 경우 스칼라값과의 계산이 넘파이 배열의 원소별로 한 번씩 수행
- 이 기능을 브로드캐스트라고 함

```python
A = np.array([[1, 2], [3, 4]])
print(A)
```

```python
A.shape
```

```python
A.dtype
```

- 행렬의 형상은 shape로

- 행렬에 담긴 원소의 자료형은 dtype으로 알 수 있음

```python
B = np.array([[3, 0], [0, 6]])
A + B
```

```python
A * B
```

- 형상이 같은 행렬끼리면 행렬의 산술 연산도 대응하는 원소별로 계산됨 (배열과 마찬가지로)

- 행렬과 스칼라값의 산술 연산도 가능 (배열과 마찬가지로 브로드캐스트 기능 작동)

```python
print(A)
```

```python
A * 10
```

- 넘파이 배열은 N차우너 배열을 작성할 수 있음

- 1차원 배열, 2차원 배열, 3차원 배열처럼 원하는 차수의 배열을 만들 수 있음
- 수학에서는 1차원 배열은 벡터(vector)
- 2차원 배열은 행렬(matrics)이라고 부름
- 벡터와 행렬을 일반화한 것을 텐서(tensor)라고 함

#### 브로드캐스트

- 넘파이에서는 형상이 다른 배열끼리도 계산 가능

- 2x2 행렬 A에 스칼라 값 10을 곱

- 10이라는 스칼라값이 2x2 행렬로 확대된 후 연산이 이루어짐
- 이러한 기능을 브로드캐스트(broadcast)라 함

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

- 1차원 배열인 B가 2차원 배열 A와 똑같은 형상으로 변형된 후 연산이 이루어짐

#### 원소 접근

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
```

```python
X[0] # 0행
```

```python
X[0][1] # (0, 1) 위치의 원소
```

```python
# for문으로 각 원소 접근
for row in X:
    print(row)
```

```python
# 인덱스를 배열로 지정해 한 번에 여러 원소에 접근
X = X.flatten() # x를 1차원 배열로 변환(평탄화)
print(X)
```

- 이 기법을 응용하면 특정 조건을 만족하는 원소만 얻기 가능

```python
# 배열 X에서 15 이상인 값만 구하기
X > 15
```

```python
X[X > 15]
```

### matplotlib

#### 단순한 그래프 그리기

- 그래프를 그리려면 matplotlib의 pyplot 모듈을 사용

```python
# sin 함수 그리기
import numpy as np 
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1간격으로
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

#### pyplot의 기능

```python
# cos 함수 그리기
# sin 함수 그리기
import numpy as np 
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1간격으로
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle = "--", label = "cod") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos') # 제목
plt.legend()
plt.show()
```
