## 계산 그래프

계산 그래프(computational graph)
- 계산 과정을 그래프로 나타낸 것

- 그래프 : 복수의 노드(node)와 에지(edge)로 표현되는 자료구조
- - 에지(edge) : 노드 사이의 직선

### 계산 그래프로 풀기

- 문제 1 : 현빈이가 슈퍼에서 1개에 100원인 사과를 2개 샀습니다. 이때 지불 금액을 구하세요. 단, 소비세가 10% 부과됩니다

- 계산 그래프는 계산 과정을 노드와 화살표로 표현함

- 노드는 원(o)으로 표기하고 원 안에 연산 내용을 적음
- 계산 결과를 화살표 위에 적어 각 노드의 계산 결과가 왼쪽에서 오른쪽으로 전해짐

- 사과의 100원이 'x2' 노드로 흘러 200원이 됨

- 그 다음 노드로 전달

- 200원이 'x1.1' 노드를 거쳐 220원아 됨

- 문제 2 : 현빈이는 슈퍼에서 사과를 2개 귤을 3개 샀습니다. 사과는 1개에 100원, 귤은 1개 150원입니다. 소비세가 10% 일 때 지불 금액을 구하세요.

- 이 문제에는 덧셈 노드인 '+'가 새로 등장하여 사과와 귤의 금액을 합산함

- 계산 그래프는 왼쪽에서 오른쪽으로 계산을 진행


- 계산 그래프의 흐름

1. 계산 그래프를 구성한다

2. 그래프에서 계산을 왼쪽에서 오른쪽으로 진행한다

- 2번째 '계산을 왼쪽에서 오른쪽으로 진행'하는 단계를 '순전파(forward propagation)'라고 함

- 순전파는 계산 그래프의 출발점부터 종착점으로의 전파
- 순전파의 반대 방향은 '역전파(backward propagation)'라고 함
- 역전파는 이후에 미분을 계산할 때 중요한 역할을 함

### 국소적 계산

- 계산 그래프의 특징은 '국소적 계산'을 전파함으로써 최종 결과를 얻는다는 점에 있음

- 국소적 : 자신과 직접 관계된 작은 범위
- 국소적 계산은 결국 전체에서 어떤 일이 벌어지든 상관없이 자신과 관계된 정보만으로 결과를 출력할 수 있음
- 각 노드는 자신과 관련한 계산 외에는 아무것도 신경 쓸 게 없음

- 계산 그래프는 국소적 계산에 집중

- 전체 계산이 아무리 복잡하더라도 각 단계에서 하는 일은 해당 노드의 국소적 계산
- 국소적은 계산은 단순하지만, 그 결과를 전달함으로써 전체를 구성하는 복잡한 계산을 해낼 수 있다

### 왜 계산 그래프로 푸는가?

- 계산 그래프의 이점

1. 국소적 계산
- 전체가 아무리 복잡해도 각 노드에서는 단순한 계산에 집중하여 문제를 단순화 할 수 있음

2. 계산 그래프는 중간 계산 결과를 모두 보관할 수 있다

3. 역전파를 통해 '미분'을 효율적으로 계산할 수 있음
- 순전파와 역전파를 활용해서 각 변수의 미분을 효율적으로 구할 수 있다


## 연쇄법칙

- 역전파는 '국소적인 미분'을 순방향과는 반대인 오른쪽에서 왼쪽으로 전달

- 이 '국소적 미분'을 전달하는 원리는 연쇄법칙(chain rule)에 따른 것


### 계산 그래프의 역전파

- 역전파의 계산 순서를 따르면 목표로 하는 미분 값을 효율적으로 구할 수 있음

- 왜 이런 일이 가능한가는 연쇄법칙의 원리로 설명할 수 있음

### 연쇄법칙이란?

- 합성 함수 : 여러 함수로 구성된 함수

- 연쇄법칙은 합성 함수의 미분에 대한 성질

- 합성 함수의 미분은 합성 함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있음 -> 연쇄법칙의 원리

### 연쇄 법칙과 계산 그래프

- 계산 그래프의 역전파는 오른쪽에서 왼쪽으로 신호를 전파

- 역전파의 계산 절차에서는 노드로 들어온 입력 신호에 그 노드의 국소적 미분(편미분)을 곱한 후 다음 노드로 전달

### 역전파

### 덧셈 노드의 역전파

- z = x + y

- ax/az와 ay/az는 모드 1

- 역전파 때는 상류에서 전해진 미분(az/al)에 1을 곱하여 하류로 흘림

- - 덧셈 노드의 역전파는 1을 곱하기만 할 뿐이므로 입력된 값을 그대로 다음 노드로 보냄

### 곱셈 노드의 역전파

- z = xy

- 이 식의 미분

- ax/az = y
- ay/az = x

- 곱셈 노드 역전파는 상류의 값에 순전파 때의 입력 신호들을 '서로 바꾼 값'을 곱해서 하류로 보냄

- - 서로 바꾼 값이란 순전파 때 x였다면 역전파에서는 y, 순전파 때 y였다면 역전파에서는 x

- 곱샘의 역전파에서는 입력 신호를 바꾼 값을 곱하여 하나는 1.3 * 5 = 6.5, 다른 하나는 1.3 * 10 = 13

- 덧셈의 역전파에서는 상류의 값을 그대로 흘려보내서 순방향 입력 신호의 값은 필요하지 않지만, 곱셈의 역전파에서는 순방향 입력 신호의 값이 필요함
- 그래서 곱셈 노드를 구현할 때는 순전파의 입력 신호를 변수에 저장해둠

### 사과 쇼핑의 예

- 이 문제에서는 사과의 가격, 사과의 개수, 소비세라는 세 변수 각각이 최종 금액에 어떻게 영향을 주느냐를 품

- 이는 '사과 가격에 대한 지불 금액의 미분', '사과 개수에 대한 지불 금액의 미분', '소비세에 대한 지불 금액의 미분'을 구하는 것에 해당

- 이를 계산 그래프의 역전파를 사용해 풀기!

- 곱셈 노드의 역전파에서는 입력 신호를 서로 바꿔서 하류로 흘림

- 사과 가격의 미분은 2.2, 사과 개수의 미분은 110, 소비세의 이분은 200
- 이는 소비세와 사과 가격이 같은 양만큼 오르면 최종 금액에는 소비세가 200의 크기로, 사과 가격이 2.2 크기로 영향을 줌
- 단, 이 예에서 소비세와 사과 가격은 단위가 다르니 주의!

## 단순한 계층 구현하기

- '사과 쇼핑'의 예를 파이썬으로 구현

- 곱셈 노드를 'MulLayer', 덧셈 노드를 'AddLayer'로 구현

### 곱셈 계층

- 모든 계층은 forward()와 backward()라는 공통의 메서드(인터페이스)를 갖도록 구현

- forwward()는 순전파, backward()는 역전파를 처리

- 곱셈 계층 구현
- MulLayer라는 이름의 클래스로 구현

```python
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None

    def forward(self, x, y):
        self.x = x
        self.y = y
        out = x * y
        return out

    def backward(self, dout):
        dx = dout * self.y  # x와 y를 바꾼다
        dy = dout * self.x  
        return dx, dy
```

- __init__()에서는 인스턴스 변수인 x와 y를 초기화 함
- 이 두 변수는 순전파 시의 입력값을 유지하기 위해서 사용

- forward()에서는 x와 y를 인수로 받고 두 값을 곱해서 반환

- 반면 backward()에서는 상류에서 넘어온 미분(dout)에 순전파 때의 값을 '서로 바꿔' 곱한 후 하류로 흘림


- 이 MulLayer를 사용해서 앞에서 본 '사과 쇼핑' 구현

```python
apple = 100
apple_num = 2
tax = 1.1

# 계층들
mul_apple_layer = MulLayer()
mul_tax_layer = MulLayer()

# 순전파
apple_price = mul_apple_layer.forward(apple, apple_num)  # 사과 가격
price = mul_tax_layer.forward(apple_price, tax)  # 세금 포함 가격

print(price)  # 220.00000000000003
```

- 각 변수에 대한 미분은 backward()에서 구할 수 있음

```python
# 역전파
dprice = 1
dapple_price, dtax = mul_tax_layer.backward(dprice)  # 세금 포함 가격에 대한 미분
dapple, dapple_num = mul_apple_layer.backward(dapple_price)  # 사과 가격에 대한 미분

print(dapple)  # 2.2
print(dapple_num)  # 110.00000000000001
print(dtax)  # 200.00000000000003
```

- backward() 호출 순서는 forward() 때와는 반대

- backward()가 받는 인수는 '순전파의 출력에 대한 미분'
- 가령 mul_apple_layer라는 곱셈 계층은 순전파 때는 apple_price를 출력, 역전파 때는 apple_price의 미분 값인 dapple_price를 인수로 받음

### 덧셈 계층

```python
class AddLayer:
    def __init__(self):
        pass

    def forward(self, x, y):
        out = x + y
        return out

    def backward(self, dout):
        dx = dout * 1  # 덧셈의 미분은 1
        dy = dout * 1
        return dx, dy
```

- 덧셈 계층에서는 초기화가 필요 없으니 __init__()에서는 아무 일도 하지 않음(pass)

- 덧셈 계층의 forward()에서는 입력받은 두 인수 x, y를 더해서 반환
- backward()에서는 상류에서 내려온 미분(dout)를 그대로 하류로 흘림

- 사과 2개와 귤 3개를 사는 상황 구현

```python
apple = 100
apple_num = 2
orange = 150
orange_num = 3
tax = 1.1

# 계층들
mul_apple_layer = MulLayer()
mul_orange_layer = MulLayer()
add_apple_orange_layer = AddLayer()
mul_tax_layer = MulLayer()

# 순전파
apple_price = mul_apple_layer.forward(apple, apple_num)  # 사과 가격
orange_price = mul_orange_layer.forward(orange, orange_num)  # 귤 가격
all_price = add_apple_orange_layer.forward(apple_price, orange_price)  # 사과와 귤 가격 합산
price = mul_tax_layer.forward(all_price, tax)  # 세금 포함 가격

print(price)  # 715.0000000000001   

# 역전파
dprice = 1
dall_price, dtax = mul_tax_layer.backward(dprice)  # 세금 포함 가격에 대한 미분
dapple_price, dorange_price = add_apple_orange_layer.backward(dall_price)  # 사과와 귤 가격 합산에 대한 미분        
dorange, dorange_num = mul_orange_layer.backward(dorange_price)  # 귤 가격에 대한 미분
dapple, dapple_num = mul_apple_layer.backward(dapple_price)  # 사과 가격에 대한 미분

print(dapple)  # 2.2
print(dapple_num)  # 110.00000000000001
print(dorange)  # 3.3000000000000003
print(dorange_num)  # 165.00000000000003
print(dtax)  # 715.0000000000001
```

- 필요한 계층을 만들어 순전파 메서드인 forward()를 적절한 순서로 호츌

- 그런 다음 순전파와 반대 순서로 역전파 메서드인 backward()를 호출하면 원하는 미분이 나옴

- 계산 그래프에서의 계층(곱셈과 덧셈)은 쉽게 구현할 수 있으며, 이를 사용해 복잡한 미분도 계산 가능
